```python
import tensorflow as tf
from tensorflow.keras import Model
from tensorflow.keras.layers import concatenate, Dense, Input
from tensorflow.keras.utils import plot_model
```

## Build the Wide and Deep Model

```python
# define the inputs
input_deep = Input(shape=[1], name='Deep_Input')
input_wide = Input(shape=[1], name='Wide_Input')

# define deep paths
hidden_1 = Dense(30, activation='relu')(input_deep)
hidden_2 = Dense(30, activation='relu')(hidden_1)

# define merged path
concat = concatenate([input_wide, hidden_2])
main_output = Dense(1)(concat)

# define another output for the deep path
aux_output = Dense(1)(hidden_2)

# build the model
model = Model(inputs=[input_deep, input_wide],
              outputs=[main_output, aux_output])

# visualze the architecture
plot_model(model)
```
